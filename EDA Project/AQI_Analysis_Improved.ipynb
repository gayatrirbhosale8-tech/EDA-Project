# Outlier Detection and Handling

```python
# Detection of Outliers using IQR method
def detect_outliers_iqr(series, label):
    Q1  = series.quantile(0.25)
    Q3  = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers    = series[(series < lower_bound) | (series > upper_bound)]
    print(f"  {label:<20}  Q1={Q1:.1f}  Q3={Q3:.1f}  IQR={IQR:.1f}"
          f"  Bounds=[{lower_bound:.1f}, {upper_bound:.1f}]"
          f"  Outliers={len(outliers):,}")
    return lower_bound, upper_bound

num_cols = ['AQI_Value', 'Avg_Temp_C', 'Humidity_Pct', 'Rainfall_mm', 'Wind_Speed_kmh']
for col in num_cols:
    detect_outliers_iqr(df[col], col)

df['AQI_Value'] = df['AQI_Value'].clip(lower=0, upper=500)
print("\n  AQI Value clipped to valid CPCB range [0, 500].")
```

# Data Binning and Formatting


```python
# Data Binning of AQI Value

aqi_bins   = [0, 50, 100, 200, 300, 400, 500]
aqi_labels = ['Good (0-50)', 'Satisfactory (51-100)',
              'Moderate (101-200)', 'Poor (201-300)',
              'Very Poor (301-400)', 'Severe (401-500)']

df['AQI_Bin'] = pd.cut(
    df['AQI_Value'],
    bins   = aqi_bins,
    labels = aqi_labels,
    right  = True,
    include_lowest = True
)
print("  AQI bin distribution:")
print(df['AQI_Bin'].value_counts().sort_index().to_string())
```

```python
# Temperature Binning

temp_bins   = [-5, 10, 20, 30, 40, 50]
temp_labels = ['Very Cold (<10°C)', 'Cold (10-20°C)',
               'Warm (20-30°C)', 'Hot (30-40°C)', 'Very Hot (>40°C)']
df['Temp_Bin'] = pd.cut(
    df['Avg_Temp_C'],
    bins   = temp_bins,
    labels = temp_labels,
    right  = True
)
print("  Temperature bin distribution:")
print(df['Temp_Bin'].value_counts().sort_index().to_string())
```

```python
# Quantile Binning

df['AQI_Quantile'] = pd.qcut(
    df['AQI_Value'],
    q      = 4,
    labels = ['Q1 Low', 'Q2 Moderate-Low', 'Q3 Moderate-High', 'Q4 High'],
    duplicates = 'drop'
)
print("  Quantile bin distribution (approx. equal record counts):")
print(df['AQI_Quantile'].value_counts().sort_index().to_string())
```

```python
# Humidity Binning

df['Humidity_Bin'] = pd.cut(
    df['Humidity_Pct'],
    bins   = [0, 40, 60, 80, 100],
    labels = ['Dry (<40%)', 'Comfortable (40-60%)',
              'Humid (60-80%)', 'Very Humid (>80%)'],
    right  = True
)
print("  Humidity bin distribution:")
print(df['Humidity_Bin'].value_counts().sort_index().to_string())
```

```python
df['date'] = pd.to_datetime(df['date'], format='%d-%m-%Y')

# Custom Column Formatting

df['Month'] = df['date'].dt.month
df['Month_Name'] = df['date'].dt.strftime('%B')
df['Day_of_Week'] = df['date'].dt.day_name()
df['Quarter'] = df['date'].dt.quarter
df['Week_No'] = df['date'].dt.isocalendar().week.astype(int)

print("  New date-part columns added: Month, Month_Name, Day_of_Week, Quarter, Week_No")
print(df[['date','Month','Month_Name','Day_of_Week','Quarter','Week_No']].head(5).to_string())
```
